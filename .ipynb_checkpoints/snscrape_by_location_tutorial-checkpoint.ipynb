```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
import itertools
```

```python
word = 'weather'
place = 'Watford, England'
loc = '51.6565, 0.3903, 10km'
n=100

```

```python
res_coord = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper('{} geocode:"{}" -filter:links -filter:replies'.format(word, loc)).get_items(), n))[['url', 'date', 'content']]

# webbrowser.open(res_coord['url'].iloc[2])

res_coord.head(10)
```

```python
res_name = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper('{} near:"{}" within:10km -filter:links -filter:replies'.format(word, place)).get_items(), n))[['url', 'date', 'content']]

# webbrowser.open(res_name['url'].iloc[2])

res_name.head(10)
```

```python
common = res_coord.merge(res_name, how='inner')

common
```
